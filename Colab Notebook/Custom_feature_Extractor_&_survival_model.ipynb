```python
import os
import numpy as np
import pandas as pd
import pydicom
from pathlib import Path
import SimpleITK as sitk
from radiomics import featureextractor
from rt_utils import RTStructBuilder
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)  # suppress radiomics warnings

# ----------------------------
# Helper: Load CT as SimpleITK image
# ----------------------------
def load_ct_as_sitk(ct_folder):
    reader = sitk.ImageSeriesReader()
    files = reader.GetGDCMSeriesFileNames(str(ct_folder))
    if not files:
        raise FileNotFoundError(f"No DICOM files found in {ct_folder}")
    reader.SetFileNames(files)
    return reader.Execute()

# ----------------------------
# Helper: Extract from DICOM-SEG
# ----------------------------
def extract_tumor_mask_from_seg(seg_file, ct_img, ct_folder):
    seg_ds = pydicom.dcmread(str(seg_file), force=True)
    dcm_files = [f for f in Path(ct_folder).rglob("*.dcm") if f.is_file()]
    ct_slices = [pydicom.dcmread(str(f), force=True) for f in dcm_files]
    ct_slices.sort(key=lambda s: float(getattr(s, 'ImagePositionPatient', [0,0,0])[2]))
    ct_sop_to_idx = {s.SOPInstanceUID: i for i, s in enumerate(ct_slices)}

    H, W = seg_ds.Rows, seg_ds.Columns
    tumor_mask_3d = np.zeros((len(ct_slices), H, W), dtype=np.uint8)

    tumor_seg_idx = None
    tumor_keywords = ["Neoplasm, Primary", "GTV", "GTV-1", "Tumor"]
    for i, seg in enumerate(seg_ds.SegmentSequence):
        label = getattr(seg, 'SegmentLabel', f"Segment-{getattr(seg, 'SegmentNumber', i+1)}")
        if any(kw in str(label) for kw in tumor_keywords):
            tumor_seg_idx = i + 1
            break

    if tumor_seg_idx is None:
        raise ValueError("Tumor label not found")

    for frame_idx in range(getattr(seg_ds, 'NumberOfFrames', 0)):
        try:
            seg_num = seg_ds.PerFrameFunctionalGroupsSequence[frame_idx].SegmentIdentificationSequence[0].ReferencedSegmentNumber
            if seg_num != tumor_seg_idx:
                continue
            ref_sop = seg_ds.PerFrameFunctionalGroupsSequence[frame_idx].DerivationImageSequence[0].SourceImageSequence[0].ReferencedSOPInstanceUID
            if ref_sop in ct_sop_to_idx:
                slice_idx = ct_sop_to_idx[ref_sop]
                frame_data = seg_ds.pixel_array[frame_idx]
                tumor_mask_3d[slice_idx] = frame_data
        except (IndexError, AttributeError, KeyError):
            continue

    mask_sitk = sitk.GetImageFromArray(tumor_mask_3d)
    mask_sitk.CopyInformation(ct_img)
    return mask_sitk

# ----------------------------
# Helper: Extract from RTSTRUCT
# ----------------------------
def extract_tumor_mask_from_rtstruct(rtstruct_file, ct_folder):
    rtstruct = RTStructBuilder.create_from(
        dicom_series_path=str(ct_folder),
        rt_struct_path=str(rtstruct_file)
    )
    roi_names = rtstruct.get_roi_names()
    target_roi = None
    for name in roi_names:
        if any(kw in name for kw in ["GTV-1", "GTV", "Tumor", "Primary"]):
            target_roi = name
            break
    if not target_roi:
        target_roi = roi_names[0]  # fallback to first ROI

    mask_3d = rtstruct.get_roi_mask_by_name(target_roi)
    mask_array = np.transpose(mask_3d.astype(np.uint8), (2, 0, 1))  # (D, H, W)

    ct_img = load_ct_as_sitk(ct_folder)
    mask_sitk = sitk.GetImageFromArray(mask_array)
    mask_sitk.CopyInformation(ct_img)
    return mask_sitk

# ----------------------------
# Main Loop: All 422 Patients
# ----------------------------
csv_path = "/content/ct_seg_paths.csv"
df = pd.read_csv(csv_path)

# Handle "NA" strings: replace with NaN, then check with pd.isna
df = df.replace("NA", np.nan)

extractor = featureextractor.RadiomicsFeatureExtractor()
results = []
total_patients = len(df)
print(f"Processing {total_patients} patients...")

for i, row in df.iterrows():
    patient_id = row["patient_id"]
    ct_folder = Path(row["ct_folder_path"])
    seg_file = row["seg_file_path"]
    rtstruct_file = Path(row["rat_file_path"])

    if i % 50 == 0:
        print(f"\nProgress: {i}/{total_patients} patients...")

    features = {"patient_id": patient_id}

    try:
        ct_img = load_ct_as_sitk(ct_folder)

        # --- SEG PATH: only if valid ---
        seg_valid = False
        if pd.notna(seg_file):
            seg_path = Path(seg_file)
            if seg_path.is_file():
                try:
                    mask_seg = extract_tumor_mask_from_seg(seg_path, ct_img, ct_folder)
                    feat = extractor.execute(ct_img, mask_seg)
                    features.update({k + "_seg": v for k, v in feat.items() if not k.startswith("diagnostics_")})
                    seg_valid = True
                except Exception as e:
                    print(f"    ⚠️ {patient_id} SEG failed: {str(e)[:100]}")
            else:
                print(f"    ⚠️ {patient_id}: SEG file not found")
        if not seg_valid:
            # Optionally log that SEG was skipped
            pass

        # --- RTSTRUCT: always attempt ---
        try:
            mask_rt = extract_tumor_mask_from_rtstruct(rtstruct_file, ct_folder)
            feat = extractor.execute(ct_img, mask_rt)
            features.update({k + "_rtstruct": v for k, v in feat.items() if not k.startswith("diagnostics_")})
        except Exception as e:
            print(f"    ❌ {patient_id} RTSTRUCT failed: {str(e)[:100]}")
            # Even if RTSTRUCT fails, we might still have SEG → still append
            if not any(k.endswith('_seg') for k in features.keys() if k != "patient_id"):
                continue  # skip if both failed

        results.append(features)

    except Exception as e:
        print(f"    💥 {patient_id} overall error: {str(e)[:100]}")
        continue

# ----------------------------
# Final Save
# ----------------------------
if not results:
    raise ValueError("No patients succeeded!")

feature_df = pd.DataFrame(results)
feature_df.set_index("patient_id", inplace=True)

output_path = "radiomics_features_422_patients_seg_rtstruct.csv"
feature_df.to_csv(output_path)
print(f"\n✅ Done! Saved features for {len(feature_df)} patients to '{output_path}'.")

# Summary
seg_cols = [c for c in feature_df.columns if c.endswith('_seg')]
rt_cols = [c for c in feature_df.columns if c.endswith('_rtstruct')]
print(f"\nFeature counts:")
print(f"- Patients with SEG features:   {feature_df[seg_cols].notna().any(axis=1).sum() if seg_cols else 0}")
print(f"- Patients with RTSTRUCT features: {feature_df[rt_cols].notna().any(axis=1).sum() if rt_cols else 0}")
print(f"- Total radiomics features (SEG + RTSTRUCT): {len(feature_df.columns) - 1}")
```

```python
import os
import numpy as np
import pandas as pd
import pydicom
from pathlib import Path
import SimpleITK as sitk
from radiomics import featureextractor
from rt_utils import RTStructBuilder
import warnings

# Suppress specific radiomics warnings to keep output clean
warnings.filterwarnings("ignore", category=RuntimeWarning)
warnings.filterwarnings("ignore", category=UserWarning)

# ----------------------------
# 1. RESEARCH-GRADE CONFIGURATION (IBSI Standard)
# ----------------------------
# This specific config makes your results reproducible and publishable.
params = {
    'binWidth': 25,                  # Standard for CT (Hounsfield Units)
    'resampledPixelSpacing': [1, 1, 1], # Isotropic resampling (Crucial!)
    'interpolator': sitk.sitkBSpline,   # High-quality interpolation
    'enableCE': True,                   # Calculate Cluster Prominence/Shade
    'force2D': False,                   # 3D Analysis
    'symmetricalGLCM': True,
    'label': 1,                         # Default label to look for in mask
}

# Initialize Extractor
extractor = featureextractor.RadiomicsFeatureExtractor(**params)

# --- ENABLE FILTERS (The Secret to High Accuracy) ---
extractor.enableImageTypeByName('Original')
extractor.enableImageTypeByName('Wavelet') # Captures Multi-scale Texture
extractor.enableImageTypeByName('LoG', customArgs={'sigma': [1.0, 3.0]}) # Edge Detection

# --- ENABLE FEATURES ---
extractor.enableFeatureClassByName('shape')
extractor.enableFeatureClassByName('firstorder')
extractor.enableFeatureClassByName('glcm')
extractor.enableFeatureClassByName('glrlm')
extractor.enableFeatureClassByName('glszm')
extractor.enableFeatureClassByName('gldm')
extractor.enableFeatureClassByName('ngtdm')

print("✅ Radiomics Extractor configured with IBSI standards (Resampling + Filters enabled).")

# ----------------------------
# 2. HELPER FUNCTIONS
# ----------------------------
def load_ct_as_sitk(ct_folder):
    reader = sitk.ImageSeriesReader()
    files = reader.GetGDCMSeriesFileNames(str(ct_folder))
    if not files:
        raise FileNotFoundError(f"No DICOM files found in {ct_folder}")
    reader.SetFileNames(files)
    return reader.Execute()

def extract_tumor_mask_from_seg(seg_file, ct_img, ct_folder):
    seg_ds = pydicom.dcmread(str(seg_file), force=True)
    dcm_files = [f for f in Path(ct_folder).rglob("*.dcm") if f.is_file()]

    # Robust slice sorting
    ct_slices = [pydicom.dcmread(str(f), force=True) for f in dcm_files]
    ct_slices.sort(key=lambda s: float(getattr(s, 'ImagePositionPatient', [0,0,0])[2]))
    ct_sop_to_idx = {s.SOPInstanceUID: i for i, s in enumerate(ct_slices)}

    H, W = seg_ds.Rows, seg_ds.Columns
    tumor_mask_3d = np.zeros((len(ct_slices), H, W), dtype=np.uint8)

    tumor_seg_idx = None
    # Common keywords for Lung Cancer Tumor GTV
    tumor_keywords = ["Neoplasm", "Primary", "GTV", "GTV-1", "Tumor", "GTVp"]

    for i, seg in enumerate(seg_ds.SegmentSequence):
        label = getattr(seg, 'SegmentLabel', f"Segment-{getattr(seg, 'SegmentNumber', i+1)}")
        if any(kw.lower() in str(label).lower() for kw in tumor_keywords):
            tumor_seg_idx = i + 1
            break

    if tumor_seg_idx is None:
        # Fallback: Use the first segment if specific name not found
        tumor_seg_idx = 1
        # raise ValueError("Tumor label not found") # Uncomment if you want to be strict

    for frame_idx in range(getattr(seg_ds, 'NumberOfFrames', 0)):
        try:
            func_group = seg_ds.PerFrameFunctionalGroupsSequence[frame_idx]
            seg_num = func_group.SegmentIdentificationSequence[0].ReferencedSegmentNumber
            if seg_num != tumor_seg_idx:
                continue
            ref_sop = func_group.DerivationImageSequence[0].SourceImageSequence[0].ReferencedSOPInstanceUID
            if ref_sop in ct_sop_to_idx:
                slice_idx = ct_sop_to_idx[ref_sop]
                frame_data = seg_ds.pixel_array[frame_idx]
                tumor_mask_3d[slice_idx] = frame_data
        except (IndexError, AttributeError, KeyError):
            continue

    mask_sitk = sitk.GetImageFromArray(tumor_mask_3d)
    mask_sitk.CopyInformation(ct_img)
    return mask_sitk

def extract_tumor_mask_from_rtstruct(rtstruct_file, ct_folder):
    rtstruct = RTStructBuilder.create_from(
        dicom_series_path=str(ct_folder),
        rt_struct_path=str(rtstruct_file)
    )
    roi_names = rtstruct.get_roi_names()
    target_roi = None

    # Prioritize GTV
    for name in roi_names:
        if any(kw in name.upper() for kw in ["GTV-1", "GTV", "TUMOR", "PRIMARY"]):
            target_roi = name
            break
    if not target_roi:
        target_roi = roi_names[0]  # fallback

    mask_3d = rtstruct.get_roi_mask_by_name(target_roi)
    mask_array = np.transpose(mask_3d.astype(np.uint8), (2, 0, 1))  # (D, H, W)

    ct_img = load_ct_as_sitk(ct_folder)
    mask_sitk = sitk.GetImageFromArray(mask_array)
    mask_sitk.CopyInformation(ct_img)
    return mask_sitk

# ----------------------------
# 3. MAIN EXTRACTION LOOP
# ----------------------------
csv_path = "/content/ct_seg_paths.csv"
df = pd.read_csv(csv_path)
df = df.replace("NA", np.nan)

results = []
total_patients = len(df)
print(f"\n🚀 Starting Extraction for {total_patients} patients...")

for i, row in df.iterrows():
    patient_id = row["patient_id"]
    ct_folder = Path(row["ct_folder_path"])
    seg_file = row["seg_file_path"]
    rtstruct_file = Path(row["rat_file_path"])

    if i % 10 == 0:
        print(f"Progress: {i}/{total_patients}...")

    features = {"patient_id": patient_id}

    try:
        ct_img = load_ct_as_sitk(ct_folder)

        # --- A. Extract from SEG (DICOM Segmentation) ---
        seg_valid = False
        if pd.notna(seg_file):
            seg_path = Path(seg_file)
            if seg_path.is_file():
                try:
                    mask_seg = extract_tumor_mask_from_seg(seg_path, ct_img, ct_folder)
                    # Run Extractor
                    feat = extractor.execute(ct_img, mask_seg)
                    # Store with suffix
                    features.update({k + "_seg": v for k, v in feat.items() if not k.startswith("diagnostics_")})
                    seg_valid = True
                except Exception as e:
                    # print(f"    ⚠️ {patient_id} SEG failed: {e}") # Optional: Uncomment for debug
                    pass

        # --- B. Extract from RTSTRUCT (Raw Contours) ---
        # Always attempt RTSTRUCT even if SEG failed (Dual-Stream Strategy)
        try:
            mask_rt = extract_tumor_mask_from_rtstruct(rtstruct_file, ct_folder)
            feat = extractor.execute(ct_img, mask_rt)
            features.update({k + "_rtstruct": v for k, v in feat.items() if not k.startswith("diagnostics_")})
        except Exception as e:
             # print(f"    ⚠️ {patient_id} RTSTRUCT failed: {e}")
             pass

        # Only append if at least one succeeded
        if len(features) > 1:
            results.append(features)

    except Exception as e:
        print(f"    💥 {patient_id} Critical Error: {e}")
        continue

# ----------------------------
# 4. SAVE RESULTS
# ----------------------------
if not results:
    raise ValueError("No patients succeeded! Check file paths.")

feature_df = pd.DataFrame(results)
output_path = "radiomics_features_IBSI_compliant.csv"
feature_df.to_csv(output_path, index=False)

print(f"\n✅ Extraction Complete!")
print(f"   Saved to: {output_path}")
print(f"   Total Patients Processed: {len(feature_df)}")
print(f"   Total Features per Patient: {len(feature_df.columns) - 1}")
```

```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter
from lifelines.utils import concordance_index
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# ----------------------------
# 1. Load Radiomics Features
# ----------------------------
radiomics_df = pd.read_csv("/kaggle/working/radiomics_features_422_patients_seg_rtstruct.csv", index_col="patient_id")

# Keep only RTSTRUCT features (more clinically reliable)
rt_cols = [col for col in radiomics_df.columns if col.endswith('_rtstruct')]
if not rt_cols:
    raise ValueError("No RTSTRUCT features found!")

# Clean column names: remove '_rtstruct' suffix
radiomics_clean = radiomics_df[rt_cols].rename(columns=lambda x: x.replace('_rtstruct', ''))
print(f"Loaded {radiomics_clean.shape[0]} patients, {radiomics_clean.shape[1]} RTSTRUCT radiomics features.")

# ----------------------------
# 2. Load Clinical Data
# ----------------------------
clinical_df = pd.read_csv("/kaggle/input/clinical-data-of-nsclc-lungi1/NSCLC-Radiomics-Lung1.clinical-version3-Oct-2019.csv")  # <-- ⚠️ UPDATE THIS PATH

# Ensure PatientID matches radiomics index (e.g., "LUNG1-001")
clinical_df = clinical_df.rename(columns={"PatientID": "patient_id"})
clinical_df.set_index("patient_id", inplace=True)

# Keep only needed columns
clinical_cols = [
    "age", "clinical.T.Stage", "Clinical.N.Stage", "Clinical.M.Stage",
    "Overall.Stage", "Histology", "gender", "Survival.time", "deadstatus.event"
]
clinical_df = clinical_df[clinical_cols].copy()

print(f"Clinical data: {clinical_df.shape[0]} patients.")

# ----------------------------
# 3. Merge Radiomics + Clinical
# ----------------------------
merged_df = radiomics_clean.join(clinical_df, how="inner")
print(f"After merging: {merged_df.shape[0]} patients with both radiomics and clinical data.")

# ----------------------------
# 4. Clean Data: Remove NA in critical columns
# ----------------------------
# Critical outcome & time variables must be present
merged_df = merged_df.dropna(subset=["Survival.time", "deadstatus.event"])

# Optional: remove patients with missing age or stage (adjust as needed)
merged_df = merged_df.dropna(subset=["age", "Overall.Stage"])

# Ensure event is integer (0/1)
merged_df["deadstatus.event"] = merged_df["deadstatus.event"].astype(int)
merged_df["Survival.time"] = merged_df["Survival.time"].astype(float)

print(f"After removing NA in key columns: {merged_df.shape[0]} patients.")

# ----------------------------
# 5. Encode Categorical Variables
# ----------------------------
# Gender: assume "M"/"F" → 1/0
if "gender" in merged_df.columns:
    merged_df["gender_M"] = (merged_df["gender"] == "M").astype(int)
    merged_df = merged_df.drop("gender", axis=1)

# Histology: one-hot encode (e.g., "adenocarcinoma", "squamous", etc.)
if "Histology" in merged_df.columns:
    histology_dummies = pd.get_dummies(merged_df["Histology"], prefix="Histology", drop_first=True)
    merged_df = pd.concat([merged_df.drop("Histology", axis=1), histology_dummies], axis=1)

# T/N/M/Overall Stage: convert to numeric if possible (e.g., "T1" → 1)
def stage_to_numeric(stage):
    if pd.isna(stage):
        return np.nan
    # Extract first number (e.g., "T2" → 2, "III" → 3)
    stage = str(stage).strip().upper()
    if stage.startswith('T'):
        stage = stage[1:]
    elif stage.startswith('N'):
        stage = stage[1:]
    elif stage.startswith('M'):
        stage = stage[1:]
    # Handle Roman numerals (simplified)
    roman_map = {"I": 1, "II": 2, "III": 3, "IV": 4}
    if stage in roman_map:
        return roman_map[stage]
    try:
        return float(stage)
    except:
        return np.nan

for col in ["clinical.T.Stage", "Clinical.N.Stage", "Clinical.M.Stage", "Overall.Stage"]:
    if col in merged_df.columns:
        merged_df[col + "_num"] = merged_df[col].apply(stage_to_numeric)
        merged_df = merged_df.drop(col, axis=1)

# Final NA removal (after encoding)
final_df = merged_df.dropna()
print(f"Final dataset after encoding & cleaning: {final_df.shape[0]} patients, {final_df.shape[1]} features.")

# ----------------------------
# 6. Cox Proportional Hazards Model
# ----------------------------
# Define duration and event columns
duration_col = "Survival.time"
event_col = "deadstatus.event"

# Prepare data for Cox model
cox_df = final_df.copy()

# Optional: standardize radiomics features (recommended for regularization)
radiomics_feature_names = [col for col in cox_df.columns if not col.startswith(("age", "clinical", "deadstatus", "Survival"))]
cox_df[radiomics_feature_names] = (cox_df[radiomics_feature_names] - cox_df[radiomics_feature_names].mean()) / cox_df[radiomics_feature_names].std()

# Fit Cox model
cph = CoxPHFitter(penalizer=0.1)  # L2 penalty to handle high-dimensional features
try:
    cph.fit(cox_df, duration_col=duration_col, event_col=event_col, show_progress=True)
except Exception as e:
    print("Model failed. Trying with fewer features (top 50 most variable radiomics)...")
    # Fallback: use only top 50 radiomics + clinical
    var_radiomics = cox_df[radiomics_feature_names].var().sort_values(ascending=False).head(50).index
    fallback_cols = list(var_radiomics) + [c for c in cox_df.columns if c not in radiomics_feature_names]
    cph.fit(cox_df[fallback_cols], duration_col=duration_col, event_col=event_col, show_progress=True)

# ----------------------------
# 7. Results & Visualization
# ----------------------------
print("\n✅ Cox Model Summary:")
print(f"Concordance Index: {cph.concordance_index_:.3f}")
print(f"Partial AIC: {cph.AIC_partial_:.1f}")

# Top 10 significant features (by z-score)
summary = cph.summary
summary["abs_z"] = summary["z"].abs()
top_features = summary.sort_values("abs_z", ascending=False).head(10)
print("\nTop 10 Most Significant Features:")
print(top_features[["coef", "exp(coef)", "se(coef)", "z", "p"]])

# Plot
plt.figure(figsize=(10, 6))
cph.plot()
plt.title("Cox PH Model: Log Partial Hazard")
plt.show()

# Survival curves for high vs low risk (median split)
risk_scores = cph.predict_partial_hazard(cox_df)
median_risk = np.median(risk_scores)
high_risk = cox_df[risk_scores >= median_risk]
low_risk = cox_df[risk_scores < median_risk]

plt.figure(figsize=(10, 6))
from lifelines import KaplanMeierFitter
kmf = KaplanMeierFitter()
ax = plt.subplot(111)
kmf.fit(high_risk[duration_col], high_risk[event_col], label="High Risk")
kmf.plot_survival_function(ax=ax)
kmf.fit(low_risk[duration_col], low_risk[event_col], label="Low Risk")
kmf.plot_survival_function(ax=ax)
plt.title("Kaplan-Meier: High vs Low Radiomics-Clinical Risk")
plt.legend()
plt.show()
```

## survival model CoxPHF

```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter, KaplanMeierFitter
from lifelines.utils import concordance_index
from lifelines.statistics import logrank_test
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import shap
import warnings
warnings.filterwarnings("ignore")

# ----------------------------
# 1. Load & Prepare Data (same as before)
# ----------------------------
radiomics_df = pd.read_csv("/content/radiomics_features.csv").set_index("patient_id")
clinical_df = pd.read_csv("/content/NSCLC-Radiomics-Lung1.clinical-version3-Oct-2019.csv")
clinical_df = clinical_df.rename(columns={"PatientID": "patient_id"}).set_index("patient_id")
survival_df = clinical_df[["Survival.time", "deadstatus.event"]]

df = radiomics_df.join(survival_df, how="inner").dropna(subset=["Survival.time", "deadstatus.event"])
print(f" Merged dataset: {len(df)} patients")

# Feature columns
duration_col = "Survival.time"
event_col = "deadstatus.event"
feature_cols = [col for col in df.columns if col not in [duration_col, event_col, "ct_path", "seg_path", "ct_path.1", "seg_path.1"]]

# Impute & clean
df[feature_cols] = df[feature_cols].fillna(df[feature_cols].median())
zero_var = df[feature_cols].columns[df[feature_cols].var() == 0]
if len(zero_var) > 0:
    print(f" Dropping zero-variance: {list(zero_var)}")
    df = df.drop(columns=zero_var)
    feature_cols = [c for c in feature_cols if c not in zero_var]

df = df.dropna()
print(f"➡️ Final: {len(df)} patients, {len(feature_cols)} features")

# Split & scale
X = df[feature_cols]
y_time = df[duration_col]
y_event = df[event_col]

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=feature_cols, index=X.index)

X_train, X_test, y_t_train, y_t_test, y_e_train, y_e_test = train_test_split(
    X_scaled, y_time, y_event, test_size=0.2, stratify=y_event, random_state=42
)

train_df = pd.concat([X_train.reset_index(drop=True),
                      pd.Series(y_t_train.values, name=duration_col),
                      pd.Series(y_e_train.values, name=event_col)], axis=1)
test_df = pd.concat([X_test.reset_index(drop=True),
                     pd.Series(y_t_test.values, name=duration_col),
                     pd.Series(y_e_test.values, name=event_col)], axis=1)

# ----------------------------
# 2. Fit CoxPH Model
# ----------------------------
cph = CoxPHFitter(penalizer=0.1)
cph.fit(train_df, duration_col=duration_col, event_col=event_col)

# ----------------------------
# 3. 📊 Full Evaluation Report
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY COXPH MODEL: FULL EVALUATION")
print("="*60)

# A. Model Summary
print("\n📈 MODEL SUMMARY (Top 10 Features by |coef|):")
summary = cph.summary
summary_sorted = summary.reindex(summary['coef'].abs().sort_values(ascending=False).index)
print(summary_sorted[['coef', 'exp(coef)', 'p']].head(10))

# B. C-index
train_c = cph.concordance_index_
partial_hazards_test = cph.predict_partial_hazard(test_df).values.ravel()
test_c = concordance_index(test_df[duration_col], -partial_hazards_test, test_df[event_col])
print(f"\n✅ Concordance Index:")
print(f"   - Training: {train_c:.3f}")
print(f"   - Test:     {test_c:.3f}")

# C. Significant Features (p < 0.05)
sig_features = summary[summary['p'] < 0.05].index.tolist()
print(f"\n🔬 Significant Features (p < 0.05):")
if sig_features:
    for feat in sig_features:
        hr = summary.loc[feat, 'exp(coef)']
        direction = "↑ risk" if hr > 1 else "↓ risk"
        print(f"   - {feat}: HR = {hr:.2f} ({direction})")
else:
    print("   - None")

# ----------------------------
# 4. 📈 Kaplan-Meier Curves
# ----------------------------
median_risk = np.median(partial_hazards_test)
test_df_plot = test_df.copy()
test_df_plot['risk'] = partial_hazards_test
high = test_df_plot[partial_hazards_test >= median_risk]
low = test_df_plot[partial_hazards_test < median_risk]

kmf = KaplanMeierFitter()
plt.figure(figsize=(8, 5))
kmf.fit(high[duration_col], high[event_col], label="High Risk")
kmf.plot_survival_function()
kmf.fit(low[duration_col], low[event_col], label="Low Risk")
kmf.plot_survival_function()

plt.title(f"Radiomics Risk Groups (C-index = {test_c:.3f})")
plt.xlabel("Days")
plt.ylabel("Survival Probability")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Log-rank test
lr = logrank_test(high[duration_col], low[duration_col], high[event_col], low[event_col])
print(f"\n🧮 Log-rank Test: p = {lr.p_value:.4f}")

# ----------------------------
# 5. 🌟 SHAP Values (for Interpretability)
# ----------------------------
print("\n🎨 Computing SHAP values...")

# For CoxPH (linear model), SHAP values = coefficients * (feature_value - mean)
# We can compute them manually or use shap.LinearExplainer
coefs = cph.params_.values  # shape: (n_features,)
X_test_array = X_test.values

# Center features (SHAP for linear models uses mean-centered features)
X_mean = X_train.mean().values  # baseline = training mean
shap_values = (X_test_array - X_mean) * coefs  # shape: (n_samples, n_features)

# Create SHAP summary plot
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values, X_test, feature_names=feature_cols, plot_type="dot", show=False)
plt.title("SHAP Feature Importance (Radiomics-Only CoxPH)")
plt.tight_layout()
plt.show()

# Top 10 features by mean |SHAP|
mean_abs_shap = np.mean(np.abs(shap_values), axis=0)
top_idx = np.argsort(mean_abs_shap)[::-1][:10]
print("\n🔝 Top 10 Features by Mean |SHAP| (Global Impact):")
for i in top_idx:
    print(f"   - {feature_cols[i]}: {mean_abs_shap[i]:.4f}")
```

```python
# ----------------------------
# 6. ➗ CoxNet (Regularized Cox Regression)
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY COXNET MODEL")
print("="*60)

# Import required modules
from sksurv.linear_model import CoxnetSurvivalAnalysis
from sksurv.util import Surv
from sksurv.metrics import concordance_index_censored
import shap

# Prepare survival labels in sksurv format
y_train_sksurv = Surv.from_arrays(event=y_e_train.astype(bool), time=y_t_train)
y_test_sksurv = Surv.from_arrays(event=y_e_test.astype(bool), time=y_t_test)

# Initialize and fit CoxNet
# Adjusting l1_ratio to allow for feature selection (ElasticNet with a stronger L1 component)
coxnet = CoxnetSurvivalAnalysis(
    l1_ratio=0.7,          # Increased from 1e-9 to 0.7 to encourage sparsity/feature selection
    alpha_min_ratio=0.01,
    n_alphas=100,
    normalize=False,       # already standardized
    max_iter=2000
)
coxnet.fit(X_train.values, y_train_sksurv)

# Select best alpha by training C-index
cindex_path = []
for i, alpha in enumerate(coxnet.alphas_):
    preds = coxnet.predict(X_train.values, alpha=alpha)
    cindex = concordance_index_censored(y_train_sksurv['event'], y_train_sksurv['time'], preds)[0]
    cindex_path.append(cindex)

best_idx = np.argmax(cindex_path)
best_alpha = coxnet.alphas_[best_idx]
print(f"✅ Best alpha: {best_alpha:.6f} (Train C-index: {cindex_path[best_idx]:.3f})")

# Get coefficients for best alpha
coefs_coxnet = coxnet.coef_[:, best_idx]  # shape: (n_features,)
print(f"✅ Non-zero coefficients: {np.sum(coefs_coxnet != 0)} / {len(coefs_coxnet)}")

# Predict on test set
risk_coxnet = coxnet.predict(X_test.values, alpha=best_alpha)
cindex_coxnet = concordance_index_censored(
    y_test_sksurv['event'],
    y_test_sksurv['time'],
    risk_coxnet
)[0]
print(f"\n➔ CoxNet Test C-index: {cindex_coxnet:.3f}")

# ----------------------------
# SHAP for CoxNet (exact linear SHAP)
# ----------------------------
print("\n✒ Computing SHAP values for CoxNet...")

# Center features around training mean (baseline)
X_test_array = X_test.values
X_train_mean = X_train.mean().values  # shape: (n_features,)

# SHAP = coef * (x - x_mean)
shap_values_coxnet = (X_test_array - X_train_mean) * coefs_coxnet  # (n_samples, n_features)

# Plot SHAP summary
plt.figure(figsize=(10, 8))
shap.summary_plot(
    shap_values_coxnet,
    X_test,
    feature_names=feature_cols,
    plot_type="dot",
    show=False
)
plt.title(f"SHAP: CoxNet (C-index = {cindex_coxnet:.3f})")
plt.tight_layout()
plt.show()

# Top 10 features by mean |SHAP|
mean_abs_shap = np.mean(np.abs(shap_values_coxnet), axis=0)
top_idx = np.argsort(mean_abs_shap)[::-1][:10]
print("\n↑ Top 10 Features by Mean |SHAP| (CoxNet):")
for i in top_idx:
    print(f"   - {feature_cols[i]}: {mean_abs_shap[i]:.4f}")

# ----------------------------
# Optional: Compare to CoxPH
# ----------------------------
print(f"\n✨ Comparison:")
print(f"   - CoxPH (lifelines) Test C-index: {test_c:.3f}")
print(f"   - CoxNet (sksurv)     Test C-index: {cindex_coxnet:.3f}")
```

#### DeepSurv (PyTorch) + SHAP

```python
# ----------------------------
# 8. 🧠 DeepSurv (Deep Cox) + SHAP
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY DEEPSURV")
print("="*60)

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import shap

# Prepare tensors
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
X_train_t = torch.tensor(X_train.values, dtype=torch.float32).to(device)
X_test_t = torch.tensor(X_test.values, dtype=torch.float32).to(device)
y_time_t = torch.tensor(y_t_train.values, dtype=torch.float32).to(device)
y_event_t = torch.tensor(y_e_train.values, dtype=torch.float32).to(device)

# DeepSurv model
class DeepSurv(nn.Module):
    def __init__(self, input_dim, layers=[64, 32], dropout=0.3):
        super().__init__()
        layers = [input_dim] + layers
        self.net = nn.Sequential()
        for i in range(len(layers) - 1):
            self.net.append(nn.Linear(layers[i], layers[i+1]))
            self.net.append(nn.ReLU())
            self.net.append(nn.Dropout(dropout))
        self.risk = nn.Linear(layers[-1], 1)
    def forward(self, x):
        return self.risk(self.net(x)).squeeze()

model = DeepSurv(X_train.shape[1]).to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)

# Cox partial likelihood loss
def cox_loss(risk, times, events):
    # Sort by time descending
    idx = torch.argsort(times, descending=True)
    risk = risk[idx]
    events = events[idx]
    # Compute log cumulative sum of exp(risk)
    cumsum_exp = torch.cumsum(torch.exp(risk), dim=0)
    loss = -torch.sum(events * (risk - torch.log(cumsum_exp))) / torch.sum(events)
    return loss

# Training loop
train_loader = DataLoader(TensorDataset(X_train_t, y_time_t, y_event_t), batch_size=32, shuffle=True)
model.train()
for epoch in range(300):
    total_loss = 0
    for x, t, e in train_loader:
        optimizer.zero_grad()
        risk_pred = model(x)
        loss = cox_loss(risk_pred, t, e)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss/len(train_loader):.4f}")

# Evaluate
model.eval()
with torch.no_grad():
    risk_deepsurv = model(X_test_t).cpu().numpy()
cindex_deepsurv = concordance_index(y_t_test, -risk_deepsurv, y_e_test)
print(f"🎯 DeepSurv Test C-index: {cindex_deepsurv:.3f}")

# ----------------------------
# SHAP for DeepSurv (GradientSHAP)
# ----------------------------
print("\n✒ Computing SHAP values for DeepSurv...")

# Move model to CPU for SHAP compatibility
model_cpu = model.cpu()
X_train_np = X_train.values.astype('float32')
X_test_np = X_test.values.astype('float32')

# Use GradientExplainer (requires background reference)
background = shap.sample(X_train_np, min(100, len(X_train_np)))
explainer_ds = shap.GradientExplainer(model_cpu, background)
shap_values_ds = explainer_ds.shap_values(X_test_np)

# Summary plot
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values_ds, X_test, feature_names=feature_cols, plot_type="dot", show=False)
plt.title(f"SHAP: DeepSurv (C-index = {cindex_deepsurv:.3f})")
plt.tight_layout()
plt.show()

# Top 10 features
mean_abs_shap_ds = np.mean(np.abs(shap_values_ds), axis=0)
top_idx_ds = np.argsort(mean_abs_shap_ds)[::-1][:10]
print("\n↑ Top 10 Features by Mean |SHAP| (DeepSurv):")
for i in top_idx_ds:
    print(f"   - {feature_cols[i]}: {mean_abs_shap_ds[i]:.4f}")
```

```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter, KaplanMeierFitter
from lifelines.utils import concordance_index
from lifelines.statistics import logrank_test
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import shap
import warnings
warnings.filterwarnings("ignore")

# ----------------------------
# 1. Load & Prepare Data (same as before)
# ----------------------------
radiomics_df = pd.read_csv("/content/radiomics_features_422_patients_seg_rtstruct.csv").set_index("patient_id")
clinical_df = pd.read_csv("/content/NSCLC-Radiomics-Lung1.clinical-version3-Oct-2019.csv")
clinical_df = clinical_df.rename(columns={"PatientID": "patient_id"}).set_index("patient_id")
survival_df = clinical_df[["Survival.time", "deadstatus.event"]]

df = radiomics_df.join(survival_df, how="inner").dropna(subset=["Survival.time", "deadstatus.event"])
print(f" Merged dataset: {len(df)} patients")

# Feature columns
duration_col = "Survival.time"
event_col = "deadstatus.event"
feature_cols = [col for col in df.columns if col not in [duration_col, event_col]]

# Impute & clean
df[feature_cols] = df[feature_cols].fillna(df[feature_cols].median())
zero_var = df[feature_cols].columns[df[feature_cols].var() == 0]
if len(zero_var) > 0:
    print(f" Dropping zero-variance: {list(zero_var)}")
    df = df.drop(columns=zero_var)
    feature_cols = [c for c in feature_cols if c not in zero_var]

df = df.dropna()
print(f"➡️ Final: {len(df)} patients, {len(feature_cols)} features")

# Split & scale
X = df[feature_cols]
y_time = df[duration_col]
y_event = df[event_col]

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=feature_cols, index=X.index)

X_train, X_test, y_t_train, y_t_test, y_e_train, y_e_test = train_test_split(
    X_scaled, y_time, y_event, test_size=0.2, stratify=y_event, random_state=42
)

train_df = pd.concat([X_train.reset_index(drop=True),
                      pd.Series(y_t_train.values, name=duration_col),
                      pd.Series(y_e_train.values, name=event_col)], axis=1)
test_df = pd.concat([X_test.reset_index(drop=True),
                     pd.Series(y_t_test.values, name=duration_col),
                     pd.Series(y_e_test.values, name=event_col)], axis=1)

# ----------------------------
# 2. Fit CoxPH Model
# ----------------------------
cph = CoxPHFitter(penalizer=0.1)
cph.fit(train_df, duration_col=duration_col, event_col=event_col)

# ----------------------------
# 3. 📊 Full Evaluation Report
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY COXPH MODEL: FULL EVALUATION")
print("="*60)

# A. Model Summary
print("\n📈 MODEL SUMMARY (Top 10 Features by |coef|):")
summary = cph.summary
summary_sorted = summary.reindex(summary['coef'].abs().sort_values(ascending=False).index)
print(summary_sorted[['coef', 'exp(coef)', 'p']].head(10))

# B. C-index
train_c = cph.concordance_index_
partial_hazards_test = cph.predict_partial_hazard(test_df).values.ravel()
test_c = concordance_index(test_df[duration_col], -partial_hazards_test, test_df[event_col])
print(f"\n✅ Concordance Index:")
print(f"   - Training: {train_c:.3f}")
print(f"   - Test:     {test_c:.3f}")

# C. Significant Features (p < 0.05)
sig_features = summary[summary['p'] < 0.05].index.tolist()
print(f"\n🔬 Significant Features (p < 0.05):")
if sig_features:
    for feat in sig_features:
        hr = summary.loc[feat, 'exp(coef)']
        direction = "↑ risk" if hr > 1 else "↓ risk"
        print(f"   - {feat}: HR = {hr:.2f} ({direction})")
else:
    print("   - None")

# ----------------------------
# 4. 📈 Kaplan-Meier Curves
# ----------------------------
median_risk = np.median(partial_hazards_test)
test_df_plot = test_df.copy()
test_df_plot['risk'] = partial_hazards_test
high = test_df_plot[partial_hazards_test >= median_risk]
low = test_df_plot[partial_hazards_test < median_risk]

kmf = KaplanMeierFitter()
plt.figure(figsize=(8, 5))
kmf.fit(high[duration_col], high[event_col], label="High Risk")
kmf.plot_survival_function()
kmf.fit(low[duration_col], low[event_col], label="Low Risk")
kmf.plot_survival_function()

plt.title(f"Radiomics Risk Groups (C-index = {test_c:.3f})")
plt.xlabel("Days")
plt.ylabel("Survival Probability")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Log-rank test
lr = logrank_test(high[duration_col], low[duration_col], high[event_col], low[event_col])
print(f"\n🧮 Log-rank Test: p = {lr.p_value:.4f}")

# ----------------------------
# 5. 🌟 SHAP Values (for Interpretability)
# ----------------------------
print("\n🎨 Computing SHAP values...")

# For CoxPH (linear model), SHAP values = coefficients * (feature_value - mean)
# We can compute them manually or use shap.LinearExplainer
coefs = cph.params_.values  # shape: (n_features,)
X_test_array = X_test.values

# Center features (SHAP for linear models uses mean-centered features)
X_mean = X_train.mean().values  # baseline = training mean
shap_values = (X_test_array - X_mean) * coefs  # shape: (n_samples, n_features)

# Create SHAP summary plot
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values, X_test, feature_names=feature_cols, plot_type="dot", show=False)
plt.title("SHAP Feature Importance (Radiomics-Only CoxPH)")
plt.tight_layout()
plt.show()

# Top 10 features by mean |SHAP|
mean_abs_shap = np.mean(np.abs(shap_values), axis=0)
top_idx = np.argsort(mean_abs_shap)[::-1][:10]
print("\n🔝 Top 10 Features by Mean |SHAP| (Global Impact):")
for i in top_idx:
    print(f"   - {feature_cols[i]}: {mean_abs_shap[i]:.4f}")
```

```python
# ----------------------------
# 6. ➗ CoxNet (Regularized Cox Regression)
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY COXNET MODEL")
print("="*60)

# Import required modules
from sksurv.linear_model import CoxnetSurvivalAnalysis
from sksurv.util import Surv
from sksurv.metrics import concordance_index_censored
import shap

# Prepare survival labels in sksurv format
y_train_sksurv = Surv.from_arrays(event=y_e_train.astype(bool), time=y_t_train)
y_test_sksurv = Surv.from_arrays(event=y_e_test.astype(bool), time=y_t_test)

# Initialize and fit CoxNet
# Adjusting l1_ratio to allow for feature selection (ElasticNet with a stronger L1 component)
coxnet = CoxnetSurvivalAnalysis(
    l1_ratio=0.7,          # Increased from 1e-9 to 0.7 to encourage sparsity/feature selection
    alpha_min_ratio=0.01,
    n_alphas=100,
    normalize=False,       # already standardized
    max_iter=2000
)
coxnet.fit(X_train.values, y_train_sksurv)

# Select best alpha by training C-index
cindex_path = []
for i, alpha in enumerate(coxnet.alphas_):
    preds = coxnet.predict(X_train.values, alpha=alpha)
    cindex = concordance_index_censored(y_train_sksurv['event'], y_train_sksurv['time'], preds)[0]
    cindex_path.append(cindex)

best_idx = np.argmax(cindex_path)
best_alpha = coxnet.alphas_[best_idx]
print(f"✅ Best alpha: {best_alpha:.6f} (Train C-index: {cindex_path[best_idx]:.3f})")

# Get coefficients for best alpha
coefs_coxnet = coxnet.coef_[:, best_idx]  # shape: (n_features,)
print(f"✅ Non-zero coefficients: {np.sum(coefs_coxnet != 0)} / {len(coefs_coxnet)}")

# Predict on test set
risk_coxnet = coxnet.predict(X_test.values, alpha=best_alpha)
cindex_coxnet = concordance_index_censored(
    y_test_sksurv['event'],
    y_test_sksurv['time'],
    risk_coxnet
)[0]
print(f"\n➔ CoxNet Test C-index: {cindex_coxnet:.3f}")

# ----------------------------
# SHAP for CoxNet (exact linear SHAP)
# ----------------------------
print("\n✒ Computing SHAP values for CoxNet...")

# Center features around training mean (baseline)
X_test_array = X_test.values
X_train_mean = X_train.mean().values  # shape: (n_features,)

# SHAP = coef * (x - x_mean)
shap_values_coxnet = (X_test_array - X_train_mean) * coefs_coxnet  # (n_samples, n_features)

# Plot SHAP summary
plt.figure(figsize=(10, 8))
shap.summary_plot(
    shap_values_coxnet,
    X_test,
    feature_names=feature_cols,
    plot_type="dot",
    show=False
)
plt.title(f"SHAP: CoxNet (C-index = {cindex_coxnet:.3f})")
plt.tight_layout()
plt.show()

# Top 10 features by mean |SHAP|
mean_abs_shap = np.mean(np.abs(shap_values_coxnet), axis=0)
top_idx = np.argsort(mean_abs_shap)[::-1][:10]
print("\n↑ Top 10 Features by Mean |SHAP| (CoxNet):")
for i in top_idx:
    print(f"   - {feature_cols[i]}: {mean_abs_shap[i]:.4f}")

# ----------------------------
# Optional: Compare to CoxPH
# ----------------------------
print(f"\n✨ Comparison:")
print(f"   - CoxPH (lifelines) Test C-index: {test_c:.3f}")
print(f"   - CoxNet (sksurv)     Test C-index: {cindex_coxnet:.3f}")
```

## Random Survival Forest (RSF) + SHAP

```python
# ----------------------------
# 7.  Random Survival Forest (RSF) + SHAP
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY RANDOM SURVIVAL FOREST")
print("="*60)

from sksurv.ensemble import RandomSurvivalForest
import shap

# Train RSF
rsf = RandomSurvivalForest(
    n_estimators=200,
    min_samples_split=5,
    min_samples_leaf=3,
    max_features="sqrt",
    random_state=42,
    n_jobs=-1
)
rsf.fit(X_train.values, y_train_sksurv)

# Predict risk: use negative of cumulative hazard at median time
# Alternative: integrate or use mean survival time
surv_funcs = rsf.predict_survival_function(X_test.values)
# Use median observed time as evaluation timepoint
median_time = np.median(y_t_test)
risk_rsf = np.array([fn(median_time) for fn in surv_funcs])  # survival prob at median_time
# Higher risk = lower survival probability → use -risk_rsf for C-index
cindex_rsf = concordance_index_censored(
    y_test_sksurv['event'],
    y_test_sksurv['time'],
    -risk_rsf  # negative because lower survival prob = higher risk
)[0]
print(f"🎯 RSF Test C-index: {cindex_rsf:.3f}")

# ----------------------------
# SHAP for RSF (TreeSHAP)
# ----------------------------
print("\n✒ Computing SHAP values for RSF...")
print("   (Using KernelExplainer, this may take a while...)")

# Initialize SHAP explainer using KernelExplainer for model-agnostic interpretation
# KernelExplainer requires a prediction function and a background dataset.
# We use a subset of the training data as the background dataset for speed.
explainer_rsf = shap.KernelExplainer(rsf.predict, shap.sample(X_train, 50).values)
shap_values_rsf = explainer_rsf.shap_values(X_test.values)

# Summary plot
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values_rsf, X_test, feature_names=feature_cols, plot_type="dot", show=False)
plt.title(f"SHAP: RSF (C-index = {cindex_rsf:.3f})")
plt.tight_layout()
plt.show()

# Top 10 features by mean |SHAP|
mean_abs_shap_rsf = np.mean(np.abs(shap_values_rsf), axis=0)
top_idx_rsf = np.argsort(mean_abs_shap_rsf)[::-1][:10]
print("\n↑ Top 10 Features by Mean |SHAP| (RSF):")
for i in top_idx_rsf:
    print(f"   - {feature_cols[i]}: {mean_abs_shap_rsf[i]:.4f}")
```

## DeepSurv (PyTorch) + SHAP

```python
# ----------------------------
# 8. 🧠 DeepSurv (Deep Cox) + SHAP
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY DEEPSURV")
print("="*60)

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import shap

# Prepare tensors
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
X_train_t = torch.tensor(X_train.values, dtype=torch.float32).to(device)
X_test_t = torch.tensor(X_test.values, dtype=torch.float32).to(device)
y_time_t = torch.tensor(y_t_train.values, dtype=torch.float32).to(device)
y_event_t = torch.tensor(y_e_train.values, dtype=torch.float32).to(device)

# DeepSurv model
class DeepSurv(nn.Module):
    def __init__(self, input_dim, layers=[64, 32], dropout=0.3):
        super().__init__()
        layers = [input_dim] + layers
        self.net = nn.Sequential()
        for i in range(len(layers) - 1):
            self.net.append(nn.Linear(layers[i], layers[i+1]))
            self.net.append(nn.ReLU())
            self.net.append(nn.Dropout(dropout))
        self.risk = nn.Linear(layers[-1], 1)
    def forward(self, x):
        return self.risk(self.net(x)).squeeze()

model = DeepSurv(X_train.shape[1]).to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)

# Cox partial likelihood loss
def cox_loss(risk, times, events):
    # Sort by time descending
    idx = torch.argsort(times, descending=True)
    risk = risk[idx]
    events = events[idx]
    # Compute log cumulative sum of exp(risk)
    cumsum_exp = torch.cumsum(torch.exp(risk), dim=0)
    loss = -torch.sum(events * (risk - torch.log(cumsum_exp))) / torch.sum(events)
    return loss

# Training loop
train_loader = DataLoader(TensorDataset(X_train_t, y_time_t, y_event_t), batch_size=32, shuffle=True)
model.train()
for epoch in range(300):
    total_loss = 0
    for x, t, e in train_loader:
        optimizer.zero_grad()
        risk_pred = model(x)
        loss = cox_loss(risk_pred, t, e)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss/len(train_loader):.4f}")

# Evaluate
model.eval()
with torch.no_grad():
    risk_deepsurv = model(X_test_t).cpu().numpy()
cindex_deepsurv = concordance_index(y_t_test, -risk_deepsurv, y_e_test)
print(f"🎯 DeepSurv Test C-index: {cindex_deepsurv:.3f}")

# ----------------------------
# SHAP for DeepSurv (GradientSHAP)
# ----------------------------
print("\n✒ Computing SHAP values for DeepSurv...")

# Move model to CPU for SHAP compatibility
model_cpu = model.cpu()
X_train_np = X_train.values.astype('float32')
X_test_np = X_test.values.astype('float32')

# Use GradientExplainer (requires background reference)
background = shap.sample(X_train_np, min(100, len(X_train_np)))
explainer_ds = shap.GradientExplainer(model_cpu, background)
shap_values_ds = explainer_ds.shap_values(X_test_np)

# Summary plot
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values_ds, X_test, feature_names=feature_cols, plot_type="dot", show=False)
plt.title(f"SHAP: DeepSurv (C-index = {cindex_deepsurv:.3f})")
plt.tight_layout()
plt.show()

# Top 10 features
mean_abs_shap_ds = np.mean(np.abs(shap_values_ds), axis=0)
top_idx_ds = np.argsort(mean_abs_shap_ds)[::-1][:10]
print("\n↑ Top 10 Features by Mean |SHAP| (DeepSurv):")
for i in top_idx_ds:
    print(f"   - {feature_cols[i]}: {mean_abs_shap_ds[i]:.4f}")
```
