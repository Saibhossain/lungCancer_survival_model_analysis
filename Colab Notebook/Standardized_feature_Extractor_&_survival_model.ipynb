```python

from pathlib import Path
import pandas as pd

DATA_ROOT = Path("/root/.cache/kagglehub/datasets/umutkrdrms/nsclc-radiomics/versions/2/NSCLC-Radiomics")
OUTPUT_CSV = "ct_seg_paths.csv"
MIN_CT_FILES = 10
records = []

for study_dir in sorted(DATA_ROOT.glob("LUNG1-*/*")):
    if not study_dir.is_dir():
        continue

    patient_id = study_dir.parent.name
    seg_files = []
    for item in study_dir.iterdir():
        if item.is_dir() and item.name.startswith("300."):
            seg_dcms = list(item.rglob("*.dcm"))
            seg_files.extend(seg_dcms)

    seg_file_path = str(seg_files[0]) if seg_files else "NA"
    best_ct_dir = None
    max_ct_count = 0
    rat_file_path = None

    for item in study_dir.iterdir():
        if not (item.is_dir() and not item.name.startswith("300.")):
            continue

        ct_dcms = list(item.rglob("*.dcm"))
        if len(ct_dcms) >= MIN_CT_FILES and len(ct_dcms) > max_ct_count:
            max_ct_count = len(ct_dcms)
            best_ct_dir = item
        if len(ct_dcms) == 1 and rat_file_path is None:
            rat_file_path = str(ct_dcms[0])

    if best_ct_dir is None:
        print(f" No valid CT folder (≥{MIN_CT_FILES} files) for {patient_id}")
        continue

    records.append({
        "patient_id": patient_id,
        "study_folder": study_dir.name,
        "ct_folder_path": str(best_ct_dir),
        "seg_file_path": seg_file_path,
        "rat_file_path": rat_file_path
    })

df = pd.DataFrame(records)
df.to_csv(OUTPUT_CSV, index=False)

print(f"\n Saved {len(df)} patient records to '{OUTPUT_CSV}'")
print("\nFirst few rows:")
print(df.head())
```

# feature extraction pipeline

```python
# features_extractor.py
import os, warnings
from pathlib import Path
import numpy as np
import pandas as pd
import pydicom
import SimpleITK as sitk

warnings.filterwarnings("ignore")

# ----------------------------
# Utils
# ----------------------------
def voxel_volume_mm3(img: sitk.Image):
    sx, sy, sz = img.GetSpacing()
    return float(sx*sy*sz)

def mask_bbox(mask_np):
    idx = np.argwhere(mask_np > 0)
    if idx.size == 0:
        return None
    zmin,ymin,xmin = idx.min(0)
    zmax,ymax,xmax = idx.max(0) + 1
    return (int(zmin),int(zmax),int(ymin),int(ymax),int(xmin),int(xmax))

def percentile_safe(a, q):
    if a.size == 0:
        return np.nan
    return np.percentile(a, q)

# ----------------------------
# Feature extraction (fallback / PyRadiomics optional)
# ----------------------------
def extract_features_fallback(ct_img: sitk.Image, mask_img: sitk.Image, patient_id: str):
    feats = {"patient_id": patient_id}

    # align mask to CT if needed
    if (ct_img.GetSize()!=mask_img.GetSize()) or (ct_img.GetSpacing()!=mask_img.GetSpacing()) \
       or (ct_img.GetOrigin()!=mask_img.GetOrigin()) or (ct_img.GetDirection()!=mask_img.GetDirection()):
        mask_img = sitk.Resample(mask_img, ct_img, sitk.Transform(),
                                 sitk.sitkNearestNeighbor, 0, sitk.sitkUInt8)

    ct = sitk.GetArrayFromImage(ct_img).astype(np.float32)  # [z,y,x]
    m  = sitk.GetArrayFromImage(mask_img).astype(np.uint8)
    vsize = voxel_volume_mm3(ct_img)

    # crop bbox for speed
    bb = mask_bbox(m)
    if bb is not None:
        z0,z1,y0,y1,x0,x1 = bb
        ct = ct[z0:z1, y0:y1, x0:x1]
        m  =  m[z0:z1, y0:y1, x0:x1]

    tumor = ct[m>0]
    feats["voxels"] = int(m.sum())
    feats["volume_mm3"] = feats["voxels"] * vsize

    # intensity stats (HU)
    if tumor.size > 0:
        tmin, tmax = float(np.min(tumor)), float(np.max(tumor))
        if not np.isfinite(tmin) or not np.isfinite(tmax) or tmax <= tmin:
            tmin, tmax = tmin, tmin + 1.0

        feats.update({
            "hu_mean": float(np.mean(tumor)),
            "hu_std" : float(np.std(tumor)),
            "hu_min" : float(np.min(tumor)),
            "hu_max" : float(np.max(tumor)),
            "hu_p10" : float(percentile_safe(tumor,10)),
            "hu_p25" : float(percentile_safe(tumor,25)),
            "hu_p50" : float(percentile_safe(tumor,50)),
            "hu_p75" : float(percentile_safe(tumor,75)),
            "hu_p90" : float(percentile_safe(tumor,90)),
        })

        # Shannon entropy over 64 bins
        hist, _ = np.histogram(tumor, bins=64, range=(tmin, tmax))
        if hist.sum() > 0:
            p = hist.astype(np.float64) / hist.sum()
            p = p[p > 0]
            feats["hu_entropy"] = float(-(p * np.log(p)).sum())
        else:
            feats["hu_entropy"] = np.nan
    else:
        for k in ["hu_mean","hu_std","hu_min","hu_max","hu_p10","hu_p25","hu_p50","hu_p75","hu_p90","hu_entropy"]:
            feats[k] = np.nan

    # simple shape proxies (slice extents)
    feats["extent_z"] = float(np.count_nonzero(m.any(axis=(1,2))))
    feats["extent_y"] = float(np.count_nonzero(m.any(axis=(0,2))))
    feats["extent_x"] = float(np.count_nonzero(m.any(axis=(0,1))))

    # 2D GLCM over central ~5 tumor slices (optional)
    try:
        from skimage.feature import graycomatrix, graycoprops
        slice_has = np.where(m.any(axis=(1,2)))[0]
        if slice_has.size > 0:
            mid = slice_has.size//2
            sl_idx = slice_has[max(0, mid-2):min(slice_has.size, mid+3)]
        else:
            sl_idx = []

        props_accum = []
        for k in sl_idx:
            sl = ct[k]; slm = m[k]
            if slm.sum() < 100:
                continue
            roi = sl[slm>0]
            lo, hi = np.percentile(roi, (1, 99))
            if hi <= lo: hi = lo + 1.0
            slc = np.clip(sl, lo, hi)
            q = np.floor((slc - lo) / (hi-lo) * 63.0).astype(np.uint8)

            glcm = graycomatrix(q, distances=[1,2], angles=[0,np.pi/4,np.pi/2,3*np.pi/4],
                                levels=64, symmetric=True, normed=True)
            props = {}
            for nm in ["contrast","dissimilarity","homogeneity","ASM","energy","correlation"]:
                val = np.mean(graycoprops(glcm, nm))
                props[nm] = float(val)
            props_accum.append(props)

        if props_accum:
            keys = props_accum[0].keys()
            for k in keys:
                feats[f"glcm_{k}"] = float(np.mean([d[k] for d in props_accum]))
        else:
            for nm in ["contrast","dissimilarity","homogeneity","ASM","energy","correlation"]:
                feats[f"glcm_{nm}"] = np.nan
    except Exception:
        for nm in ["contrast","dissimilarity","homogeneity","ASM","energy","correlation"]:
            feats[f"glcm_{nm}"] = np.nan

    return feats

# ----------------------------
# CT/SEG loading (file-structure aware)
# ----------------------------
def load_ct_img_from_study_dir(study_dir: Path):
    """
    Find and load the CT image series folder (usually starts with '1.' or similar).
    Returns (sitk.Image, series_uid) or (None, None) if not found.
    """
    ct_folder = None
    for sub in sorted(os.listdir(study_dir)):
        if sub.startswith("1.") or ("-NA-" in sub and not sub.startswith("3.")):
            candidate = study_dir / sub
            if not candidate.is_dir():
                continue
            dcm_files = [f for f in os.listdir(candidate) if f.lower().endswith(".dcm")]
            if len(dcm_files) > 10:
                ct_folder = candidate
                break
    if ct_folder is None:
        return None, None

    reader = sitk.ImageSeriesReader()
    series_files = reader.GetGDCMSeriesFileNames(str(ct_folder))
    reader.SetFileNames(series_files)
    ct_img = reader.Execute()

    ds = pydicom.dcmread(series_files[0], stop_before_pixels=True, force=True)
    series_uid = getattr(ds, "SeriesInstanceUID", None)
    return sitk.Cast(ct_img, sitk.sitkFloat32), series_uid

def build_mask_from_seg(study_dir: Path, ct_img: sitk.Image):
    """
    Build a tumor-only mask directly in CT grid by painting SEG frames into
    the nearest CT slice. Works for scalar multi-segment SEGs.
    """
    import numpy as np, pydicom, os, SimpleITK as sitk

    # 1) find segmentation file in 300.*
    seg_file = None
    for sub in sorted(os.listdir(study_dir)):
        if sub.startswith("300."):
            candidate = study_dir / sub
            if candidate.is_dir():
                for f in os.listdir(candidate):
                    if f.lower().endswith(".dcm"):
                        seg_file = candidate / f
                        break
    if seg_file is None:
        print("[SEG] No segmentation file found.")
        return None

    ds = pydicom.dcmread(str(seg_file), force=True)

    # 2) list segments and choose tumor-like SegmentNumber(s)
    seg_defs = []
    for s in ds.SegmentSequence:
        num = int(getattr(s, "SegmentNumber", 1))
        label = getattr(s, "SegmentLabel", "")
        cat = ""
        try:
            cat = getattr(s.SegmentedPropertyCategoryCodeSequence[0], "CodeMeaning", "")
        except Exception:
            pass
        seg_defs.append((num, label, cat))

    #print("[SEG] Found segments:", len(seg_defs))
    # for num, lab, cat in seg_defs:
    #     print(f"       {num} - {lab} | Category: {cat}")

    KEY = ("gtv", "neoplasm", "primary", "tumor")
    tumor_nums = [num for (num, lab, cat) in seg_defs
                  if any(k in lab.lower() for k in KEY) or any(k in cat.lower() for k in KEY)]
    if not tumor_nums:
        # fallback later to most frequent non-zero segment
        tumor_nums = None

    # 3) precompute CT slice physical z positions
    size = ct_img.GetSize()  # (x,y,z)
    ct_zs = np.array([ct_img.TransformIndexToPhysicalPoint((0, 0, k))[2]
                      for k in range(size[2])], dtype=np.float64)

    def nearest_ct_k(z_world: float) -> int:
        return int(np.argmin(np.abs(ct_zs - z_world)))

    pf = ds.PerFrameFunctionalGroupsSequence
    if not pf or len(pf) == 0:
        print("[SEG] No PerFrameFunctionalGroupsSequence; returning empty mask.")
        empty = sitk.Image(ct_img.GetSize(), sitk.sitkUInt8); empty.CopyInformation(ct_img)
        return empty

    rows = int(getattr(ds, "Rows", 0))
    cols = int(getattr(ds, "Columns", 0))

    # 4) determine which SegmentNumber to keep (if unspecified)
    if tumor_nums is None:
        counts = {}
        for i in range(len(pf)):
            seg_id_seq = pf[i].get("SegmentIdentificationSequence", None)
            if not seg_id_seq: continue
            num = int(seg_id_seq[0].ReferencedSegmentNumber)
            counts[num] = counts.get(num, 0) + 1
        if counts:
            tumor_nums = [max(counts.items(), key=lambda kv: kv[1])[0]]  # most frequent
        else:
            tumor_nums = []

    # 5) paint into a CT-shaped 3D mask
    mask_np = np.zeros((size[2], size[1], size[0]), dtype=np.uint8)  # [z,y,x] like sitk array

    for i in range(len(pf)):
        pfg = pf[i]
        seg_id_seq = pfg.get("SegmentIdentificationSequence", None)
        if not seg_id_seq:
            continue
        seg_num = int(seg_id_seq[0].ReferencedSegmentNumber)
        if tumor_nums and seg_num not in tumor_nums:
            continue

        pos_seq = pfg.get("PlanePositionSequence", None)
        if pos_seq is None:
            continue
        z_world = float(pos_seq[0].ImagePositionPatient[2])
        k = nearest_ct_k(z_world)
        if k < 0 or k >= size[2]:
            continue

        frame2d = ds.pixel_array[i]  # 0/255
        # If needed, check size match; usually (rows, cols) == (size[1], size[0])
        if frame2d.shape != (size[1], size[0]):
            # Simple conservative fix: skip mismatched frames (rare in this dataset)
            # Optionally, resize to CT in-plane size if you encounter mismatches
            continue

        mask_np[k] = np.maximum(mask_np[k], (frame2d > 0).astype(np.uint8))

    # 6) convert to image in CT geometry
    seg_mask = sitk.GetImageFromArray(mask_np)  # [z,y,x]
    seg_mask.CopyInformation(ct_img)            # <-- sizes now match exactly
    return seg_mask

# ----------------------------
# Driver to build feature table
# ----------------------------
def build_feature_table(dataset_root, max_patients=None, out_csv="radiomics_features.csv"):
    DATA_ROOT = Path(dataset_root)
    study_dirs = sorted({
    p for pattern in ["LUNG1-*/*StudyID*", "LUNG1-*/*NA-NA*"]
    for p in DATA_ROOT.glob(pattern)})
    if max_patients:
        study_dirs = study_dirs[:max_patients]

    rows = []

    # try PyRadiomics
    have_pyr = False
    try:
        import radiomics  # noqa
        have_pyr = True
    except Exception:
        pass

    for study_dir in study_dirs:
        patient_id = study_dir.parent.name
        print(f"\n=== {patient_id} ===")

        # CT
        try:
            ct_img, _ = load_ct_img_from_study_dir(study_dir)
            if ct_img is None:
                print("No usable CT series found.")
                continue
        except Exception as e:
            print(f"[WARN] {patient_id}: CT load error:", e)
            continue
        print("CT loaded:", ct_img.GetSize(), "spacing:", ct_img.GetSpacing())

        # SEG → mask
        try:
            mask_img = build_mask_from_seg(study_dir, ct_img)
            if mask_img is None:
                print("No usable SEG mask found.")
                continue
        except Exception as e:
            print(f"[WARN] {patient_id}: SEG error:", e)
            continue
        vox = int(sitk.GetArrayFromImage(mask_img).sum())
        print(f"SEG mask built; voxels={vox}")

        # features
        try:
            if have_pyr:
                from radiomics import featureextractor
                feats = extract_features_pyradiomics(ct_img, mask_img, patient_id=None)  # optional
            feats = extract_features_fallback(ct_img, mask_img, patient_id)
            feats["ct_path"] = str(study_dir)  # add CT folder path
            feats["seg_path"] = str(study_dir)  # add segmentation folder path (same base study_dir)
            rows.append(feats)
        except Exception as e:
            print(f"[WARN] {patient_id}: feature extraction error:", e)

    df = pd.DataFrame(rows)
    nun = df.isna().mean()
    keep = [c for c in df.columns if c=="patient_id" or (nun[c] < 0.7)]
    df = df[keep + ["ct_path", "seg_path"]]
    df.to_csv(out_csv, index=False)
    print(f"\nSaved features to: {out_csv}")
    return df

# (Optional) PyRadiomics richer features (kept here if you enable it)
def extract_features_pyradiomics(ct_img: sitk.Image, mask_img: sitk.Image, patient_id: str):
    from radiomics import featureextractor
    params = {
        "imageType": {"Original": {}},
        "featureClass": {"firstorder": [], "shape": [], "glcm": [], "glrlm": [], "glszm": [], "ngtdm": []},
        "setting": {"binWidth": 25, "resampledPixelSpacing": None, "interpolator": "sitkBSpline", "normalize": False}
    }
    extractor = featureextractor.RadiomicsFeatureExtractor(params)
    mask_img = sitk.Cast(mask_img>0, sitk.sitkUInt8)
    result = extractor.execute(ct_img, mask_img)

    feats = {"patient_id": patient_id}
    for k, v in result.items():
        if k.startswith("diagnostics_"): continue
        try: feats[k] = float(v)
        except: pass
    return feats

if __name__ == "__main__":
    # Example CLI-like usage inside notebooks:
    DATASET_ROOT = "/root/.cache/kagglehub/datasets/umutkrdrms/nsclc-radiomics/versions/2/NSCLC-Radiomics"
    build_feature_table(DATASET_ROOT, max_patients=None, out_csv="radiomics_features.csv")
```

## survival model CoxPHF

```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter, KaplanMeierFitter
from lifelines.utils import concordance_index
from lifelines.statistics import logrank_test
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import shap
import warnings
warnings.filterwarnings("ignore")

# ----------------------------
# 1. Load & Prepare Data (same as before)
# ----------------------------
radiomics_df = pd.read_csv("/content/radiomics_features.csv").set_index("patient_id")
clinical_df = pd.read_csv("/content/NSCLC-Radiomics-Lung1.clinical-version3-Oct-2019.csv")
clinical_df = clinical_df.rename(columns={"PatientID": "patient_id"}).set_index("patient_id")
survival_df = clinical_df[["Survival.time", "deadstatus.event"]]

df = radiomics_df.join(survival_df, how="inner").dropna(subset=["Survival.time", "deadstatus.event"])
print(f" Merged dataset: {len(df)} patients")

# Feature columns
duration_col = "Survival.time"
event_col = "deadstatus.event"
feature_cols = [col for col in df.columns if col not in [duration_col, event_col, "ct_path", "seg_path", "ct_path.1", "seg_path.1"]]

# Impute & clean
df[feature_cols] = df[feature_cols].fillna(df[feature_cols].median())
zero_var = df[feature_cols].columns[df[feature_cols].var() == 0]
if len(zero_var) > 0:
    print(f" Dropping zero-variance: {list(zero_var)}")
    df = df.drop(columns=zero_var)
    feature_cols = [c for c in feature_cols if c not in zero_var]

df = df.dropna()
print(f"➡️ Final: {len(df)} patients, {len(feature_cols)} features")

# Split & scale
X = df[feature_cols]
y_time = df[duration_col]
y_event = df[event_col]

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=feature_cols, index=X.index)

X_train, X_test, y_t_train, y_t_test, y_e_train, y_e_test = train_test_split(
    X_scaled, y_time, y_event, test_size=0.2, stratify=y_event, random_state=42
)

train_df = pd.concat([X_train.reset_index(drop=True),
                      pd.Series(y_t_train.values, name=duration_col),
                      pd.Series(y_e_train.values, name=event_col)], axis=1)
test_df = pd.concat([X_test.reset_index(drop=True),
                     pd.Series(y_t_test.values, name=duration_col),
                     pd.Series(y_e_test.values, name=event_col)], axis=1)

# ----------------------------
# 2. Fit CoxPH Model
# ----------------------------
cph = CoxPHFitter(penalizer=0.1)
cph.fit(train_df, duration_col=duration_col, event_col=event_col)

# ----------------------------
# 3. 📊 Full Evaluation Report
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY COXPH MODEL: FULL EVALUATION")
print("="*60)

# A. Model Summary
print("\n📈 MODEL SUMMARY (Top 10 Features by |coef|):")
summary = cph.summary
summary_sorted = summary.reindex(summary['coef'].abs().sort_values(ascending=False).index)
print(summary_sorted[['coef', 'exp(coef)', 'p']].head(10))

# B. C-index
train_c = cph.concordance_index_
partial_hazards_test = cph.predict_partial_hazard(test_df).values.ravel()
test_c = concordance_index(test_df[duration_col], -partial_hazards_test, test_df[event_col])
print(f"\n✅ Concordance Index:")
print(f"   - Training: {train_c:.3f}")
print(f"   - Test:     {test_c:.3f}")

# C. Significant Features (p < 0.05)
sig_features = summary[summary['p'] < 0.05].index.tolist()
print(f"\n🔬 Significant Features (p < 0.05):")
if sig_features:
    for feat in sig_features:
        hr = summary.loc[feat, 'exp(coef)']
        direction = "↑ risk" if hr > 1 else "↓ risk"
        print(f"   - {feat}: HR = {hr:.2f} ({direction})")
else:
    print("   - None")

# ----------------------------
# 4. 📈 Kaplan-Meier Curves
# ----------------------------
median_risk = np.median(partial_hazards_test)
test_df_plot = test_df.copy()
test_df_plot['risk'] = partial_hazards_test
high = test_df_plot[partial_hazards_test >= median_risk]
low = test_df_plot[partial_hazards_test < median_risk]

kmf = KaplanMeierFitter()
plt.figure(figsize=(8, 5))
kmf.fit(high[duration_col], high[event_col], label="High Risk")
kmf.plot_survival_function()
kmf.fit(low[duration_col], low[event_col], label="Low Risk")
kmf.plot_survival_function()

plt.title(f"Radiomics Risk Groups (C-index = {test_c:.3f})")
plt.xlabel("Days")
plt.ylabel("Survival Probability")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Log-rank test
lr = logrank_test(high[duration_col], low[duration_col], high[event_col], low[event_col])
print(f"\n🧮 Log-rank Test: p = {lr.p_value:.4f}")

# ----------------------------
# 5. 🌟 SHAP Values (for Interpretability)
# ----------------------------
print("\n🎨 Computing SHAP values...")

# For CoxPH (linear model), SHAP values = coefficients * (feature_value - mean)
# We can compute them manually or use shap.LinearExplainer
coefs = cph.params_.values  # shape: (n_features,)
X_test_array = X_test.values

# Center features (SHAP for linear models uses mean-centered features)
X_mean = X_train.mean().values  # baseline = training mean
shap_values = (X_test_array - X_mean) * coefs  # shape: (n_samples, n_features)

# Create SHAP summary plot
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values, X_test, feature_names=feature_cols, plot_type="dot", show=False)
plt.title("SHAP Feature Importance (Radiomics-Only CoxPH)")
plt.tight_layout()
plt.show()

# Top 10 features by mean |SHAP|
mean_abs_shap = np.mean(np.abs(shap_values), axis=0)
top_idx = np.argsort(mean_abs_shap)[::-1][:10]
print("\n🔝 Top 10 Features by Mean |SHAP| (Global Impact):")
for i in top_idx:
    print(f"   - {feature_cols[i]}: {mean_abs_shap[i]:.4f}")
```

```python
# ----------------------------
# 6. ➗ CoxNet (Regularized Cox Regression)
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY COXNET MODEL")
print("="*60)

# Import required modules
from sksurv.linear_model import CoxnetSurvivalAnalysis
from sksurv.util import Surv
from sksurv.metrics import concordance_index_censored
import shap

# Prepare survival labels in sksurv format
y_train_sksurv = Surv.from_arrays(event=y_e_train.astype(bool), time=y_t_train)
y_test_sksurv = Surv.from_arrays(event=y_e_test.astype(bool), time=y_t_test)

# Initialize and fit CoxNet
# Adjusting l1_ratio to allow for feature selection (ElasticNet with a stronger L1 component)
coxnet = CoxnetSurvivalAnalysis(
    l1_ratio=0.7,          # Increased from 1e-9 to 0.7 to encourage sparsity/feature selection
    alpha_min_ratio=0.01,
    n_alphas=100,
    normalize=False,       # already standardized
    max_iter=2000
)
coxnet.fit(X_train.values, y_train_sksurv)

# Select best alpha by training C-index
cindex_path = []
for i, alpha in enumerate(coxnet.alphas_):
    preds = coxnet.predict(X_train.values, alpha=alpha)
    cindex = concordance_index_censored(y_train_sksurv['event'], y_train_sksurv['time'], preds)[0]
    cindex_path.append(cindex)

best_idx = np.argmax(cindex_path)
best_alpha = coxnet.alphas_[best_idx]
print(f"✅ Best alpha: {best_alpha:.6f} (Train C-index: {cindex_path[best_idx]:.3f})")

# Get coefficients for best alpha
coefs_coxnet = coxnet.coef_[:, best_idx]  # shape: (n_features,)
print(f"✅ Non-zero coefficients: {np.sum(coefs_coxnet != 0)} / {len(coefs_coxnet)}")

# Predict on test set
risk_coxnet = coxnet.predict(X_test.values, alpha=best_alpha)
cindex_coxnet = concordance_index_censored(
    y_test_sksurv['event'],
    y_test_sksurv['time'],
    risk_coxnet
)[0]
print(f"\n➔ CoxNet Test C-index: {cindex_coxnet:.3f}")

# ----------------------------
# SHAP for CoxNet (exact linear SHAP)
# ----------------------------
print("\n✒ Computing SHAP values for CoxNet...")

# Center features around training mean (baseline)
X_test_array = X_test.values
X_train_mean = X_train.mean().values  # shape: (n_features,)

# SHAP = coef * (x - x_mean)
shap_values_coxnet = (X_test_array - X_train_mean) * coefs_coxnet  # (n_samples, n_features)

# Plot SHAP summary
plt.figure(figsize=(10, 8))
shap.summary_plot(
    shap_values_coxnet,
    X_test,
    feature_names=feature_cols,
    plot_type="dot",
    show=False
)
plt.title(f"SHAP: CoxNet (C-index = {cindex_coxnet:.3f})")
plt.tight_layout()
plt.show()

# Top 10 features by mean |SHAP|
mean_abs_shap = np.mean(np.abs(shap_values_coxnet), axis=0)
top_idx = np.argsort(mean_abs_shap)[::-1][:10]
print("\n↑ Top 10 Features by Mean |SHAP| (CoxNet):")
for i in top_idx:
    print(f"   - {feature_cols[i]}: {mean_abs_shap[i]:.4f}")

# ----------------------------
# Optional: Compare to CoxPH
# ----------------------------
print(f"\n✨ Comparison:")
print(f"   - CoxPH (lifelines) Test C-index: {test_c:.3f}")
print(f"   - CoxNet (sksurv)     Test C-index: {cindex_coxnet:.3f}")
```

## DeepSurv (PyTorch) + SHAP

```python
# ----------------------------
# 8.  DeepSurv (Deep Cox) + SHAP
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY DEEPSURV")
print("="*60)

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import shap

# Prepare tensors
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
X_train_t = torch.tensor(X_train.values, dtype=torch.float32).to(device)
X_test_t = torch.tensor(X_test.values, dtype=torch.float32).to(device)
y_time_t = torch.tensor(y_t_train.values, dtype=torch.float32).to(device)
y_event_t = torch.tensor(y_e_train.values, dtype=torch.float32).to(device)

# DeepSurv model
class DeepSurv(nn.Module):
    def __init__(self, input_dim, layers=[64, 32], dropout=0.3):
        super().__init__()
        layers = [input_dim] + layers
        self.net = nn.Sequential()
        for i in range(len(layers) - 1):
            self.net.append(nn.Linear(layers[i], layers[i+1]))
            self.net.append(nn.ReLU())
            self.net.append(nn.Dropout(dropout))
        self.risk = nn.Linear(layers[-1], 1)
    def forward(self, x):
        # Removed .squeeze() to ensure output is 2D (batch_size, 1) for SHAP
        return self.risk(self.net(x))

model = DeepSurv(X_train.shape[1]).to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)

# Cox partial likelihood loss
def cox_loss(risk, times, events):
    # Sort by time descending
    idx = torch.argsort(times, descending=True)
    risk = risk[idx]
    events = events[idx]
    # Compute log cumulative sum of exp(risk)
    cumsum_exp = torch.cumsum(torch.exp(risk), dim=0)
    loss = -torch.sum(events * (risk - torch.log(cumsum_exp))) / torch.sum(events)
    return loss

# Training loop
train_loader = DataLoader(TensorDataset(X_train_t, y_time_t, y_event_t), batch_size=32, shuffle=True)
model.train()
for epoch in range(300):
    total_loss = 0
    for x, t, e in train_loader:
        optimizer.zero_grad()
        risk_pred = model(x)
        loss = cox_loss(risk_pred, t, e)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss/len(train_loader):.4f}")

# Evaluate
model.eval()
with torch.no_grad():
    risk_deepsurv = model(X_test_t).cpu().numpy().squeeze()
cindex_deepsurv = concordance_index(y_t_test, -risk_deepsurv, y_e_test)
print(f"█ DeepSurv Test C-index: {cindex_deepsurv:.3f}")

# ----------------------------
# SHAP for DeepSurv (GradientSHAP)
# ----------------------------
print("\n✒ Computing SHAP values for DeepSurv...")

# Move model to CPU for SHAP compatibility
model_cpu = model.cpu()
X_train_np = X_train.values.astype('float32')
X_test_np = X_test.values.astype('float32')

# Use GradientExplainer (requires background reference)
background = shap.sample(X_train_np, min(100, len(X_train_np)))
explainer_ds = shap.GradientExplainer(model_cpu, torch.tensor(background, dtype=torch.float32))
shap_values_ds = explainer_ds.shap_values(torch.tensor(X_test_np, dtype=torch.float32))

# Summary plot
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values_ds, X_test, feature_names=feature_cols, plot_type="dot", show=False)
plt.title(f"SHAP: DeepSurv (C-index = {cindex_deepsurv:.3f})")
plt.tight_layout()
plt.show()

# Top 10 features
mean_abs_shap_ds = np.mean(np.abs(shap_values_ds), axis=0).squeeze() # Add .squeeze() here
top_idx_ds = np.argsort(mean_abs_shap_ds)[::-1][:10]
print("\n↑ Top 10 Features by Mean |SHAP| (DeepSurv):")
for i in top_idx_ds:
    print(f"   - {feature_cols[i]}: {mean_abs_shap_ds[i]:.4f}")
```

```python
# Improved survival modeling pipeline for radiomics + CoxPH
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from lifelines import CoxPHFitter, KaplanMeierFitter
from lifelines.utils import concordance_index, k_fold_cross_validation
from lifelines.statistics import logrank_test

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from statsmodels.stats.outliers_influence import variance_inflation_factor

import shap
import random
random.seed(42)
np.random.seed(42)

# ----------------------------
# Utility functions
# ----------------------------
def compute_vif(df, thresh=5.0):
    """
    Iteratively remove features with VIF > thresh.
    Returns reduced DataFrame and list of dropped features.
    """
    X = df.copy()
    dropped = []
    while True:
        if X.shape[1] <= 1:
            break
        vif = pd.Series(
            [variance_inflation_factor(X.values, i) for i in range(X.shape[1])],
            index=X.columns
        )
        max_vif = vif.max()
        if max_vif > thresh:
            drop_feature = vif.idxmax()
            dropped.append(drop_feature)
            X = X.drop(columns=[drop_feature])
        else:
            break
    return X, dropped

def univariate_cox_filter(df, duration_col, event_col, features, p_threshold=0.2):
    """
    Fit single-variable CoxPH for each feature and keep those with p < p_threshold.
    """
    keep = []
    for f in features:
        tmp = df[[duration_col, event_col, f]].dropna()
        if tmp.shape[0] < 10:
            continue
        cph = CoxPHFitter()
        try:
            cph.fit(tmp, duration_col=duration_col, event_col=event_col, show_progress=False)
            p = float(cph.summary.loc[f, "p"])
            if p < p_threshold:
                keep.append(f)
        except Exception:
            continue
    return keep

def bootstrap_concordance(model, X_test_df, duration_col, event_col, n_boot=200):
    """
    Compute bootstrap confidence interval for concordance on test set.
    """
    scores = []
    n = len(X_test_df)
    for _ in range(n_boot):
        idx = np.random.choice(range(n), size=n, replace=True)
        sample = X_test_df.iloc[idx]
        preds = model.predict_partial_hazard(sample).values.ravel()
        c = concordance_index(sample[duration_col], -preds, sample[event_col])
        scores.append(c)
    return np.mean(scores), np.percentile(scores, 2.5), np.percentile(scores, 97.5)

# ----------------------------
# 1. Load & Prepare Data
# ----------------------------
radiomics_df = pd.read_csv("/content/radiomics_features.csv").set_index("patient_id")
clinical_df = pd.read_csv("/content/NSCLC-Radiomics-Lung1.clinical-version3-Oct-2019.csv")
clinical_df = clinical_df.rename(columns={"PatientID": "patient_id"}).set_index("patient_id")
survival_df = clinical_df[["Survival.time", "deadstatus.event"]]

df = radiomics_df.join(survival_df, how="inner").dropna(subset=["Survival.time", "deadstatus.event"])
print(f" Merged dataset: {len(df)} patients")

# Feature columns (exclude common path-like columns if present)
duration_col = "Survival.time"
event_col = "deadstatus.event"
exclude_cols = {duration_col, event_col, "ct_path", "seg_path", "ct_path.1", "seg_path.1"}
feature_cols = [col for col in df.columns if col not in exclude_cols]

# Impute & drop zero variance features
df[feature_cols] = df[feature_cols].fillna(df[feature_cols].median())
zero_var = df[feature_cols].columns[df[feature_cols].var() == 0].tolist()
if zero_var:
    print(" Dropping zero-variance features:", zero_var)
    df = df.drop(columns=zero_var)
    feature_cols = [c for c in feature_cols if c not in zero_var]

df = df.dropna()
print(f"➡️ Final: {len(df)} patients, {len(feature_cols)} features")

# ----------------------------
# 2. Scaling & Train/Test Split
# ----------------------------
X = df[feature_cols]
y_time = df[duration_col]
y_event = df[event_col]

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=feature_cols, index=X.index)

# split (stratify by event to preserve censoring proportion)
X_train, X_test, y_t_train, y_t_test, y_e_train, y_e_test = train_test_split(
    X_scaled, y_time, y_event, test_size=0.2, stratify=y_event, random_state=42
)

# Build DataFrames that lifelines expects (we'll modify columns after filtering)
train_df_full = pd.concat([
    X_train.reset_index(drop=True),
    pd.Series(y_t_train.values, name=duration_col),
    pd.Series(y_e_train.values, name=event_col)
], axis=1)

test_df_full = pd.concat([
    X_test.reset_index(drop=True),
    pd.Series(y_t_test.values, name=duration_col),
    pd.Series(y_e_test.values, name=event_col)
], axis=1)

# ----------------------------
# 3. Feature Selection: VIF (on training set), then univariate Cox filter
# ----------------------------
print("\n🔎 Removing multicollinearity (VIF) on training data...")
X_train_for_vif = X_train.copy().reset_index(drop=True)
X_vif_reduced, dropped_vif = compute_vif(X_train_for_vif, thresh=5.0)
print(" Dropped due to VIF >", 5.0, ":", dropped_vif)
vif_features = X_vif_reduced.columns.tolist()

# Univariate Cox filtering (keep p < 0.20)
print("🔎 Univariate Cox filtering (p < 0.20)...")
uni_keep = univariate_cox_filter(
    df=pd.concat([X_train_for_vif[vif_features].reset_index(drop=True),
                  pd.Series(y_t_train.reset_index(drop=True).values, name=duration_col),
                  pd.Series(y_e_train.reset_index(drop=True).values, name=event_col)], axis=1),
    duration_col=duration_col, event_col=event_col, features=vif_features, p_threshold=0.20
)
print(" Kept after univariate filter:", len(uni_keep))
if len(uni_keep) == 0:
    # fallback: keep top 10 by variance if nothing passes
    print(" WARNING: univariate filter removed all features, falling back to top-10 variance features.")
    uni_keep = X_train_for_vif.var().sort_values(ascending=False).index[:10].tolist()

# Final feature list to use
final_features = uni_keep.copy()
print(" Final features used for modeling:", final_features)

# Prepare final train/test DataFrames with selected features
train_df = pd.concat([
    X_train_for_vif[final_features].reset_index(drop=True),
    pd.Series(y_t_train.reset_index(drop=True).values, name=duration_col),
    pd.Series(y_e_train.reset_index(drop=True).values, name=event_col)
], axis=1)

# For test, ensure same columns (reindex)
X_test_sel = X_test.reset_index(drop=True)[final_features]
test_df = pd.concat([
    X_test_sel,
    pd.Series(y_t_test.reset_index(drop=True).values, name=duration_col),
    pd.Series(y_e_test.reset_index(drop=True).values, name=event_col)
], axis=1)

# ----------------------------
# 4. Penalizer tuning via k-fold CV (on training data)
# ----------------------------
print("\n🔧 Tuning penalizer via 5-fold CV (concordance index)...")
penalties = [0.001, 0.01, 0.05, 0.1, 0.5, 1.0, 5.0]
best_pen = None
best_score = -np.inf
cv_results = {}
for p in penalties:
    cph_temp = CoxPHFitter(penalizer=p)
    try:
        scores = k_fold_cross_validation(cph_temp, train_df, duration_col=duration_col,
                                         event_col=event_col, k=5, scoring_method="concordance_index")
        mean_score = np.mean(scores)
    except Exception as e:
        mean_score = -np.inf
    cv_results[p] = mean_score
    print(f"  λ={p:>6} → mean CV C-index = {mean_score:.4f}")
    if mean_score > best_score:
        best_score = mean_score
        best_pen = p

print(f"🏆 Best penalizer (by CV): {best_pen} → mean CV C-index = {best_score:.4f}")

# ----------------------------
# 5. Fit final CoxPH with best penalizer
# ----------------------------
print("\n📦 Fitting final CoxPH model on training data...")
cph = CoxPHFitter(penalizer=best_pen)
cph.fit(train_df, duration_col=duration_col, event_col=event_col, show_progress=False)
print(" Model fitted.")

# ----------------------------
# 6. Model Summary & Concordance
# ----------------------------
print("\n" + "="*60)
print("RADIOMICS-ONLY COXPH MODEL: FULL EVALUATION")
print("="*60)

# Model summary (top 10 by |coef|)
summary = cph.summary
summary_sorted = summary.reindex(summary['coef'].abs().sort_values(ascending=False).index)
print("\n📈 MODEL SUMMARY (Top 10 Features by |coef|):")
print(summary_sorted[['coef', 'exp(coef)', 'p']].head(10))

# Concordance indices
train_preds = cph.predict_partial_hazard(train_df).values.ravel()
train_c = concordance_index(train_df[duration_col], -train_preds, train_df[event_col])

test_preds = cph.predict_partial_hazard(test_df).values.ravel()
test_c = concordance_index(test_df[duration_col], -test_preds, test_df[event_col])

print(f"\n✅ Concordance Index:")
print(f"   - Training: {train_c:.3f}")
print(f"   - Test:     {test_c:.3f}")

# Bootstrap C-index CI
mean_c, ci_low, ci_high = bootstrap_concordance(cph, test_df, duration_col, event_col, n_boot=200)
print(f"   - Test (bootstrap mean ± 95% CI): {mean_c:.3f} [{ci_low:.3f}, {ci_high:.3f}]")

# Significant features
sig_features = summary[summary['p'] < 0.05].index.tolist()
print(f"\n🔬 Significant Features (p < 0.05):")
if sig_features:
    for feat in sig_features:
        hr = summary.loc[feat, 'exp(coef)']
        direction = "↑ risk" if hr > 1 else "↓ risk"
        print(f"   - {feat}: HR = {hr:.2f} ({direction}), p = {summary.loc[feat,'p']:.4g}")
else:
    print("   - None")

# ----------------------------
# 7. Proportional Hazards assumption check
# ----------------------------
print("\n🔎 Checking proportional hazards assumption (training set).")
try:
    # This prints warnings/violations if any. set p_value_threshold as needed.
    cph.check_assumptions(train_df, show_plots=False, p_value_threshold=0.05)
except Exception as e:
    print("  check_assumptions raised an exception (some diagnostics may not be available):", e)

# ----------------------------
# 8. Kaplan-Meier curves (risk groups) + log-rank
# ----------------------------
# split test into high/low risk using median partial hazard
median_risk = np.median(test_preds)
test_df_plot = test_df.copy()
test_df_plot['partial_hazard'] = test_preds
high = test_df_plot[test_df_plot['partial_hazard'] >= median_risk]
low = test_df_plot[test_df_plot['partial_hazard'] < median_risk]

kmf = KaplanMeierFitter()
plt.figure(figsize=(8, 5))
kmf.fit(high[duration_col], high[event_col], label="High Risk")
ax = kmf.plot_survival_function()
kmf.fit(low[duration_col], low[event_col], label="Low Risk")
kmf.plot_survival_function(ax=ax)

plt.title(f"Radiomics Risk Groups (Test C-index = {test_c:.3f})")
plt.xlabel("Days")
plt.ylabel("Survival Probability")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Log-rank test
lr = logrank_test(high[duration_col], low[duration_col], high[event_col], low[event_col])
print(f"\n🧮 Log-rank Test: p = {lr.p_value:.4f}")

# ----------------------------
# 9. SHAP interpretability
# ----------------------------
print("\n🎨 Computing SHAP values (exact linear SHAP + optional Kernel SHAP verification)...")

# Exact linear SHAP (for linear Cox: contribution to linear predictor = coef * (x - mean))
coefs = cph.params_.loc[final_features].values  # aligned with final_features order
X_train_mean = train_df[final_features].mean().values
X_test_arr = test_df[final_features].values
linear_shap_values = (X_test_arr - X_train_mean) * coefs.reshape(1, -1)  # shape (n_samples, n_features)

# Global importance by mean(|SHAP|)
mean_abs_shap = np.mean(np.abs(linear_shap_values), axis=0)
shap_ranking = pd.Series(mean_abs_shap, index=final_features).sort_values(ascending=False)
print("\n🔝 Top features by mean |linear SHAP|:")
print(shap_ranking.head(10))

# SHAP summary plot (uses the computed matrix)
try:
    plt.figure(figsize=(10, 6))
    # shap.summary_plot expects either SHAP object or array; here we provide array and DataFrame
    shap.summary_plot(linear_shap_values, pd.DataFrame(test_df[final_features], columns=final_features), show=False)
    plt.title("SHAP (linear exact) summary — contributions to linear predictor")
    plt.tight_layout()
    plt.show()
except Exception as e:
    print("Could not draw SHAP summary plot (matplotlib / shap issue):", e)

# Optional: Kernel SHAP verification on a small subset (more costly). We'll run it on up-to-50 samples.
run_kernel_shap = True
if run_kernel_shap:
    try:
        subset_bg = train_df[final_features].sample(n=min(50, len(train_df)), random_state=42)
        def model_partial_hazard(X_df):
            # lifelines returns a series-like; convert to 1d numpy
            return cph.predict_partial_hazard(pd.DataFrame(X_df, columns=final_features)).values.ravel()
        print("\nRunning Kernel SHAP on a small subset (slow) — this helps verify linear SHAP patterns...")
        explainer = shap.KernelExplainer(model_partial_hazard, subset_bg, link="identity")
        subset_test = test_df[final_features].sample(n=min(50, len(test_df)), random_state=1)
        kernel_shap_vals = explainer.shap_values(subset_test, nsamples=200)
        plt.figure(figsize=(8, 6))
        shap.summary_plot(kernel_shap_vals, subset_test, show=False)
        plt.title("SHAP (Kernel) on subset — verify linear SHAP")
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(" Kernel SHAP failed or was interrupted (this is optional). Error:", e)

print("\n✅ Done. Pipeline completed.")
```
